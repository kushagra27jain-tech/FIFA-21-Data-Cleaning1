```python
def currency(value):
    value=value.replace("€","")
    if "M" in value :
      return  float(value.replace("M",""))*1_000_000
    elif "K" in value :
       return float(value.replace("K",""))*1_000
    else:
        return float(value)
```

```python
dataset1["Value"]=(dataset1["Value"].apply(currency))
```

```python
dataset1["Wage"]=(dataset1["Wage"].apply(currency))
```

```python
def incm(height):
    if "cm" in height:
        return float(height.replace("cm",""))
    if "'" in height:
        if'"' in height : 
           height=height.replace('"',"")
        feet,inch=height.split("'")
        feet = float(feet)
        inch = float(inch)
    else :
        return (height)
    return (inch*2.54+feet*30.48)
```

```python
dataset1["Height"]=dataset1["Height"].apply(incm)
```

```python
dataset1["Height"].head(500)
```

```python
def inkg(weight):
    if "kg" in weight:
        return float(weight.replace('kg',''))
    if "lbs" in weight:
        return (float(weight.replace('lbs',''))*.454)
```

```python
dataset1["Weight"]=dataset1["Weight"].apply(inkg)
```
